### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
x = np.linspace(-3, 3, 100)
y = pow(x, 3) + 1
y2 = pow(x, 2) + 1
plt.plot(x, y, label = "x^3 + 1")
plt.title("Wykresy dwóch funkcji", loc = "center")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y2, color = "green", linewidth = 2.5, label = "x^2 + 1")
plt.legend()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Wykresy przedstawione osobno')

ax1.plot(x, y, label = "x^3 + 1")
ax2.plot(x, y2, color = "green", linewidth = 2.5, label = "x^2 + 1")
ax1.set_title("Wykres funkcji x^3 + 1")
ax2.set_title("Wykres funkcji x^2 + 1")
ax1.set_xlabel("x")
ax2.set_xlabel("x")
ax1.set_ylabel("y")

plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 15, 0, 0, 0, 0, 0]


with plt.xkcd():
    plt.figure(figsize=(8,8))
    plt.plot(x,y)
    plt.yticks([])
    plt.title("Używanie wykresów kołowych przez studentów IAD")
    plt.xlabel('Semestr studiów')
    plt.ylabel('Liczba zrobionych wykresów kołowych')

    plt.annotate(
        'Techniki wizualizacji danych',
        xy=(3, 0), arrowprops=dict(arrowstyle='->'), xytext=(4, 3))
plt.show()
```
