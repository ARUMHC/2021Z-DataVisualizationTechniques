# Zadania

### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

```python
x = np.linspace(-3, 3, 100)
y = np.power(x, 3) + 1
```

```python
plt.plot(x, y)
plt.title("y = x^3 + 1", loc="center")
plt.ylabel("y")
plt.xlabel("x")
plt.show()
```

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

```python
y2 = np.exp2(x)
plt.plot(x, y, label="y = x^3 + 1")
plt.plot(x, y2, label="y = 2^x", color="red", linewidth = 5)
plt.title("Wykresy dwoch funkcji", loc="center")
plt.ylabel("y")
plt.xlabel("x")
plt.legend()
plt.show()
```

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle("Wykresy dwóch funkcji")

ax1.plot(x, y)
ax1.set_ylabel("y", rotation=0)
ax1.set_xlabel("x")

ax2.plot(x, y2, color="red", linewidth = 5)
ax2.set_ylabel("y", rotation=0)
ax2.set_xlabel("x")

plt.show()
```

# Zadanie 3

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej.
Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020.
Jestem też ciekawa Waszych innych pomysłów.

![https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png](https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png)

Źródło: https://static.im-g.pl/im/7/26498/m26498997,WOJEWODZTWA2.png

```python
wojewodztwo = np.array(["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",
               "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE",
               "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"])

l_2020 = np.array([1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665])

l_2015_2019 = np.array([1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683])

wzrost = 100*(l_2020-l_2015_2019)/l_2015_2019

sortowanie = np.argsort(wzrost)
wzrost_sorted = wzrost[sortowanie]
wojewodztwo_sorted = wojewodztwo[sortowanie]

xticks = ["0%", "10%","20%","30%","40%","50%","60%","70%","80%"]

fig, ax = plt.subplots()


ax.barh(wojewodztwo_sorted, wzrost_sorted, color="grey")
ax.set_title("Procentowy wzrost liczby zmarłych w 10.2020\n w porównaniu do średniej liczby zgonów w latach 2015-2019")
ax.set_xlabel("Procentowy wzrost")
ax.set_xticklabels(xticks)
ax.grid(axis="x", alpha=0.5)
ax.set_axisbelow(True)
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
x = [80, 20]
y = [1]
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.pie(x, startangle=45, colors=["yellow","white"])
    plt.pie(y, radius=0.1, colors=["black"], center = (0.1, 0.4))
    plt.xticks([])
    plt.yticks([])
    plt.title("When should you use pie charts?", size=30, loc='right')
    
    plt.text(-0.4,1.2,
        '1. To create pac-man', fontsize=20
        )
    plt.text(1,1.2,
        "2. And that's it.", fontsize=20
    )
    
plt.show()
```
