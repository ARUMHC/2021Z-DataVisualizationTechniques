### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

a)

```python
import numpy as np
```

```python
x = np.linspace(-3, 3, 1000)
y = x**3 + 1
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji y=x^3+1")
plt.show()
```

b)

```python
y_1 = x**2 - 2
plt.plot(x, y, label = "y=x^3+1")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykresy funkcji")
plt.plot(x, y_1, label = "y=x^2-2", linestyle = "-.", lw = 5.5)
plt.legend()
plt.show()
```

c)

```python
fig = plt.figure()
ax = plt.subplot(1, 2, 1)
ax.plot(x, y)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Wykres funkcji y=x^3+1")
bx = plt.subplot(1, 2, 2)
bx.plot(x, y, label = "y=x^3+1")
bx.set_xlabel("x")
bx.set_ylabel("y")
bx.set_title("Wykresy funkcji")
bx.plot(x, y_1, label = "y=x^2-2", linestyle = "-.", lw = 5.5)
bx.legend()
fig.tight_layout()
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
x = np.linspace(0.5, 10, 1000)
y1 = np.where(((x >= 1) & (x <= 5)), 1/x, 1)
y2 = np.where((x >= 5.5), 1/(x-4.5), 1)
y = np.where((y1 <= y2), y1, y2)


with plt.xkcd():
    plt.figure(figsize=(10,10))
    plt.plot(x,y)
    
    plt.xticks([])
    plt.yticks([])
    
    plt.title("Wykład")
    plt.xlabel('Czas')
    plt.ylabel('Moja uwaga')

    plt.annotate(
        'Kiedy usłyszę słowo\nkolokwium',
        xy=(5, 1/5), arrowprops=dict(arrowstyle='->'), xytext=(5.12, 0.24))
    plt.annotate(
        'Kiedy wydaje mi się, że wykładowca\nskończył mówić o kolokwium',
        xy=(5.5, 1), arrowprops=dict(arrowstyle='->'), xytext=(6, 0.8))
plt.show()
```
