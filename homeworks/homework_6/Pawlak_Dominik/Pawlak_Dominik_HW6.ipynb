# Zadania

### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
x = np.linspace(-3, 3, 120)
y_1 = [0] * 120
y_2 = [0] * 120
for i in range(0, 120):
    y_1[i] = (x[i]) ** 3 + 1
    y_2[i] = 2 * ((x[i] + 2) ** 2) - 3
plt.plot(x, y_1, label = 'f(x) = x^3 + 1', c = 'magenta')
plt.plot(x, y_2, label = '2x^2 + 8x + 5', c = 'black', lw = 4.5)
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.title('Wykresy przykladowych funkcji', loc = 'center')
plt.legend()
plt.show()

fig, (a, b) = plt.subplots(1, 2)
plt.suptitle('Subplots')
a.plot(x, y_1, label = 'f(x) = x^3 + 1', c = 'magenta')
b.plot(x, y_2, label = '2x^2 + 8x + 5', c = 'black', lw = 4.5)
plt.show()
```

Poprawić wykres tak, aby łatwiej było odczytać w którym województwie śmiertelność wzrosła najbardziej.
Jest przynajmniej kilka innych podejść: scatterplot z latami na osi OX/OY, odcinki zaczynające się w danych ze średniej a kończące w 2020.
Jestem też ciekawa Waszych innych pomysłów.

```python
a = {'Wojewodztwo': ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"],
    'Rok 2020': [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665],
    'Lata 2015-2019': [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]}
df_a = pd.DataFrame(a)


rok = df_a['Rok 2020']
lata = df_a['Lata 2015-2019']

przyrost = ((rok - lata)*100)/lata
for i in range(0, 16):
    przyrost[i] = round(przyrost[i], 2)

plt.barh(df_a['Wojewodztwo'], przyrost, color = 'green')
plt.xlabel('Przyrost w procentach')
plt.ylabel('Wojewodztwo')
plt.title('Procentowy przyrost liczby zgonów w październiku 2020 w stosunku do średniej w latach 2015-2019')
plt.show()


```
