```python
import matplotlib.pyplot as plt
import numpy as np
```

## Zadanie 1

```python
x = np.linspace(-3, 3, 100)
y = np.power(x, 3) + 1
plt.plot(x, y, label = "x^3+1")
plt.title('Wykres 1')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.show()
```

```python
y1 = 2*x*(np.power(x, 2)-2.5**2)
plt.plot(x, y, label = "x^3+1")
plt.plot(x, y1, label = "2x(x^2 - (2.5)^2)", linewidth = 4)
plt.legend()
plt.title('Wykres 1 i 2 razem')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
plt.title('Wykres')
ax1.plot(x, y)
ax1.title.set_text('Wykres 1')
ax1.set_xlabel('Oś X')
ax1.set_ylabel('Oś Y')
ax2.plot(x, y1, color = 'y', linewidth = 4)
ax2.title.set_text('Wykres 2')
ax2.set_xlabel('Oś X')
ax2.set_ylabel('Oś Y')
plt.subplots_adjust(bottom = -1.1)
plt.show()
```

## Zadanie 3

```python
wojewodztwo = ["ZACHODNIOPOMORSKIE", "WIELKOPOLSKIE", "WARMINSKO-MAZURSKIE", "SWIETOKRZYSKIE",  "SLASKIE", "POMORSKIE", "PODLASKIE", "PODKARPACKIE", "OPOLSKIE", "MAZOWIECKIE", "MALOPOLSKIE", "LODZKIE", "LUBUSKIE", "LUBELSKIE", "KUJAWSKO-POMORSKIE", "DOLNOSLASKIE"]

l_2020 = [1874, 4013, 1565, 2008, 6270, 2521, 1468, 2909, 1356, 6564, 4743, 3687, 1313, 2731, 2435, 3665]

l_2015_2019 = [1509, 2796, 1241, 1208, 4255, 1830, 1049, 1643, 881, 4811, 2690, 2578, 894, 1920, 1768, 2683]
```

```python
l_2020 = np.array(l_2020)
l_2015_2019 = np.array(l_2015_2019)
przyrost = (l_2020/l_2015_2019 - 1) * 100

plt.figure(figsize=(7, 5))

plt.barh(wojewodztwo, przyrost)
for index, value in enumerate(przyrost):
    plt.text(value, index-0.2, str(int(value)))

    
plt.title("Przyrost liczby zgonów w 2020 \n względem średniej w okresie 2015-2019", loc = "left")
plt.xlabel('Przyrost w %')
plt.show()
```

```python
x = np.linspace(0, 6, 100)
y = 1/(1 + np.exp(-x + 3))
y1 = 1/(1 + np.exp(x - 3))
with plt.xkcd():
    plt.plot(x, y, color = 'b')
    plt.plot(x, y1, color = 'y')
    plt.axvline(x=6, color = 'r')
    
    plt.annotate(
        'Deadline',
        xy=(6, 0.3), arrowprops=dict(arrowstyle='->', color = 'r'), xytext=(4, 0.5), color = 'r')
    
    plt.text(x=0.6, y=0.7, s='Happiness', color = 'y')
    plt.text(x=0.2, y=0.3, s='Working speed', color = 'b')

    plt.xticks([])
    plt.yticks([])
    plt.xlabel('Time')
    plt.title('Working on project', loc = 'left')
plt.show()
```
