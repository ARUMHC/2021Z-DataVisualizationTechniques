# Zadania

### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
x = np.linspace(-3, 3, 100)

plt.plot(x, x**3 + 1, label = "y = x^3 + 1", linewidth=2)
plt.plot(x, -x**2 - 2, color = "red", label = "y = -x^2 - 2", linewidth=4)
plt.legend()
plt.title("wykresy", loc = "center")
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2) # 2 wiersze, 1 kolumna
fig.suptitle('Subplot')

ax1.plot(x, x**3 + 1, label = "y = x^3 + 1", linewidth=2)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

ax2.plot(x, -x**2 - 2, color = "red", label = "y = -x^2 - 2", linewidth=4)
ax2.set_xlabel("x")
ax2.legend()

fig.suptitle("Wykresy funkcji")
plt.show()
```

# Zadanie 2

b) Znaleźć sposób (w internecie) na narysowanie wykresu kołowego na podanych danych.
 

```python
labels = ['Blue', 'Red', 'Green', 'Purple']
sizes = [15, 30, 45, 10]
```

```python
plt.bar(labels, sizes)
plt.title("bar plot")
plt.xlabel("color")
plt.ylabel("size")
plt.show()

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("pie chart", loc = "center")
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [0, 0, 0, 12, 12, 12, 12]


with plt.xkcd():
    plt.figure(figsize=(8,8))
    plt.plot(x,y)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel('Time')
    plt.ylabel('Annoyance level')
    plt.title("The annoyance of my roommates with my constant saying\n that pie charts are terrible")

    plt.annotate(
        'My start studying TWD',
        xy=(3, 0), arrowprops=dict(arrowstyle='->'), xytext=(4, 4))
plt.show()
```
