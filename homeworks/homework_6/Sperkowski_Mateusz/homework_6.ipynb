### Zadanie 1

a) Narysować wykres y=x^3+1 dla x z przedziału [-3, 3]. Dodać wycentrowany tytuł oraz opisy osi.

b) Do tego samego wykresu dodać wybraną inną funkcję i oznaczyć ją linią o innym kolorze i grubości. Dodać legendę.

c) Pokazać wykresy z a) i b) osobno używając funkcji subplot().

```python
fig, (ax1, ax2) = plt.subplots(1,2)
x = np.linspace(-3,3)
y = np.power(x,3)+1
ax1.plot(x,y, label = "x^3+1", color = "blue")
ax1.set_ylabel("x^3+1")
ax1.set_xlabel("x")
ax1.set_title("Wykres funkcji x^3+1")

ax2.plot(x,y, label = "x^3+1", color = "blue")
ax2.set_ylabel("f(x)")
ax2.set_xlabel("x")
ax2.set_title("Wykresy dwóch funkcji")
x_2 = np.linspace(-2,3,1000)
y_2 = np.log(x_2+2.000000001)+2
ax2.plot(x_2,y_2, label = "log(x+2)+2", color = "orange", linewidth = 3.0)
ax2.legend()
fig.tight_layout()
plt.show()
```

Matplotlib pozwala na proste tworzenie wykresów w stylu xkcd, służy do tego funkcja `xkcd()` z `matplotlib.pyplot`.
Warto zajrzeć do dokumentacji: https://matplotlib.org/gallery/showcase/xkcd.html
Mój przykład umieściłam poniżej.

Uwaga, w szczególności styl działa też z wykresami kołowymi i słupkowymi.

### Przykład

```python
x = [0, 0.05, 0.06, 0.9, 1]
y = [0, 0.5, 0, 0, 4]

# rysowanie w xkcd mode
with plt.xkcd():
    # ustawienie rozmiaru
    plt.figure(figsize=(10,10))
    
    plt.plot(x,y)
    
    # usuniecie podpisow osi
    plt.xticks([])
    plt.yticks([])
    
    # dodanie napisow
    plt.title("TWD project")
    plt.xlabel('Time')
    plt.ylabel('Productivity')

    plt.annotate(
        'One day untli deadline',
        xy=(0.9, 0), arrowprops=dict(arrowstyle='->'), xytext=(0.5, 1))
plt.show()
```

```python
x = np.linspace(0,2*np.pi)
y_1 = np.sin(x)
y_2 = np.cos(x)-1
with plt.xkcd():
    
    plt.plot(x, y_1, label = "Mine")
    plt.plot(x, y_2, label = "Rest of the group")
    
    plt.xticks([])
    plt.yticks([])
    
    plt.title("Motivation to work in a TWD group project")
    plt.xlabel("Time to the deadline")
    plt.ylabel("Motivation")
    plt.legend()
    plt.annotate(
        'Deadline',
        xy=(2*np.pi, 0), arrowprops=dict(arrowstyle='->'), xytext=(2*np.pi, -1))
plt.show()
```
