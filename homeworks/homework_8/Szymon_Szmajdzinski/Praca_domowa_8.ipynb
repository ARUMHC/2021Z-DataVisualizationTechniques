## Wykres poprawiony 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import matplotlib.ticker as mtick
 
# Data
r = [0,1,2,3,4,5]
raw_data = {'greenBars': [91.9, 84.8, 87.2, 84.3, 84.5, 87.8],
            'orangeBars': [8.1, 15.2, 12.8, 15.7, 15.5, 12.2]}
df = pd.DataFrame(raw_data)
 
# From raw value to percentage
totals = [i+j for i,j in zip(df['greenBars'], df['orangeBars'])]
greenBars = [i / j * 100 for i,j in zip(df['greenBars'], totals)]
orangeBars = [i / j * 100 for i,j in zip(df['orangeBars'], totals)]
 
# plot
barWidth = 0.9
names = ('Dalmacja','Istra','Zagreb','Sjeverna Hrvatska','Sredisna Hrvatska','Slavonija')
# Create green Bars
ax = plt.barh( r, greenBars, color='lightblue', edgecolor='white', height=barWidth, label="Maseczkowcy")
# Create orange Bars
bx = plt.barh(r, orangeBars, left=greenBars, color='#f9bc86', edgecolor='white', height=barWidth, label="Anty-Maseczkowcy")

# Custom x axis
plt.yticks(r, names)

rects = ax.patches

# For each bar: Place a label
for rect in rects:
    # Get X and Y placement of label from rect.
    x_value = rect.get_width() / 2 
    y_value =rect.get_y() + rect.get_height() / 2  

    # Number of points between bar and label. Change to your liking.
    space = 5
    # Vertical alignment for positive values
    ha = 'center'

    # Use X value as label and format number with one decimal place
    label = "{:.1%}".format(x_value * 2/100)

    # Create annotation
    plt.annotate(
        label,                      
        (x_value, y_value),         
        xytext=(0, 0),              
        textcoords="offset points", 
        va='center',                
        ha=ha)                      
        
rects = bx.patches

# For each bar: Place a label
for rect in rects:
    # Get X and Y placement of label from rect.
    x_value = rect.get_width() / 2 + rect.get_x()
    y_value =rect.get_y() + rect.get_height() / 2  

    # Vertical alignment for positive values
    ha = 'center'

    # Use X value as label and format number with one decimal place
    label = "{:.1%}".format((x_value - rect.get_x()) * 2/ 100)

    # Create annotation
    plt.annotate(
        label,                      
        (x_value, y_value),         
        xytext=(0, 0),              
        textcoords="offset points", 
        va='center',                
        ha=ha)   

ax1 = plt.axes()
ax1.set_frame_on(False)
ax1.axes.get_xaxis().set_visible(False)
plt.legend(loc='center', bbox_to_anchor=(0.45,-0.03), ncol=2, prop={'size': 15})
plt.gcf().set_size_inches(12, 6)
plt.title("Jaką część stanowią antymaseczkowcy w poszczególnych regionach Chorwacji", fontsize = 20)
 
# Show graphic
plt.show()
```

## Opis wykresu 

Główną wadą wykresu było zdecydowanie ucięcie osi x, która na oryginalnym wykresie zaczyna się w 75%. Taki dobór osi powoduje, że odbiorca odnosi wrażenie, że antymaseczkowców jest zdecydowanie więcej niż w rzeczywistości. Dodatkowo dodałem tytuł, aby nie było wątpliwości co prezentuje dany wykres. Uznałem również, że skoro wartości słupków są podpisane bezpośrednio na słupkach to oś x nie jest potrzebna. Zrezygnowałem również z ramki wykresu, bo również była zbędna.
